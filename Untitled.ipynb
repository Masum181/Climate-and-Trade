```python
def get_list_top_gdp_and_their_effect(data, year, secondary_perimeter, initial = "GDP (current US$)", num=10):
    filt = data['Series Name'] == initial
    top_10 = data[filt].sort_values(by=year, ascending=False)[year][:num]
    # print(top_10)
    filt = data['Series Name'] == secondary_perimeter
    sec_top_10 = data[filt].loc[top_10.index, year]
    # print(sec_top_10)
    df_top_10 = top_10.reset_index()
    # print(df_top_10.sort_values(by=year, ascending=False))
    output_df = pd.concat([top_10, sec_top_10], axis=1)
    output_df.columns = [initial, secondary_perimeter]
    # print(output_df.head())
    plt.figure(figsize=(10,4))
    ax  = plt.subplot(1,2,1)
    plt.bar(output_df.index, output_df[initial])
    plt.xticks(rotation=90)
    plt.title(initial)
    
    ax = plt.subplot(1,2,2)
    plt.bar(output_df.index, output_df[secondary_perimeter], color='yellow')
    plt.xticks(rotation=90)
    plt.title(secondary_perimeter)
    plt.show()
    
```

```python
get_list_top_gdp_and_their_effect(new_df, '2015', 'Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e)')
```

```python
def plot_of_country(df, country_name):
    country_code = df.loc[country_name, 'Country Code'].iloc[0]
    filt = (df['Country Code'] == country_code) & (df['Series Name'] == 'GDP (current US$)')
    gdp = df[filt].iloc[:, 3:].values[0]
    
    filt = (df['Country Code'] == country_code) & (df['Series Name'] == 'Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e)')
    co2_total = df[filt].iloc[:, 3:].values[0]
    # print(co2_total)

    filt = (df['Country Code'] == country_code) & (df['Series Name'] == 'Carbon dioxide (CO2) emissions excluding LULUCF per capita (t CO2e/capita)')
    co2_per_capita = df[filt].iloc[:, 3:].values[0]
    # print(co2_per_capita)

    filt = (df['Country Code'] == country_code) & (df['Series Name'] == 'Carbon intensity of GDP (kg CO2e per 2021 PPP $ of GDP)')
    carbon_intensity = df[filt].iloc[:, 3:].values[0]

    filt = (df['Country Code'] == country_code) & (df['Series Name'] == 'Total greenhouse gas emissions excluding LULUCF (Mt CO2e)')
    total_greenhouse_gas = df[filt].iloc[:, 3:].values[0]

    
    filt = (df['Country Code'] == country_code) & (df['Series Name'] == 'Foreign direct investment, net inflows (% of GDP)')
    foreign = df[filt].iloc[:, 3:].values[0]

    filt = (df['Country Code'] == country_code) & (df['Series Name'] == 'Trade (% of GDP)')
    trade = df[filt].iloc[:, 3:].values[0]
    
    filt = (df['Country Code'] == country_code) & (df['Series Name'] == 'Trade (% of GDP)')
    trade = df[filt].iloc[:, 3:].values[0]

    filt = (df['Country Code'] == country_code) & (df['Series Name'] == 'Foreign direct investment, net inflows (% of GDP)')
    foreign_direct_invest = df[filt].iloc[:, 3:].values[0]

    plt.figure(figsize=(16, 10))
    
    ax = plt.subplot(2, 2, 1)
    sns.lineplot(x=df.columns[3:], y=gdp, label='GDP', markers='o')
    sns.scatterplot(x=df.columns[3:], y=gdp, s=50, color='blue')
    for x, y in zip(df.columns[3:], gdp):
        value_in_billions = y / 1e9 # convert to billions
        text = f'{value_in_billions:.2f}B'
        ax.text(x, y + y * 0.02, text, ha='center', fontsize=8)

    ax.set_title(f"{country_name}'s GDP Over Years")
    plt.xticks(rotation=45)
    plt.xlabel('Year')
    plt.ylabel("GDP (current US$)")
    plt.legend()

    ax = plt.subplot(2,2,2)

    sns.lineplot(x = df.columns[3:], y= foreign, label='Foreign Direct Investment', markers='o')
    sns.scatterplot(x = df.columns[3:], y= foreign, s=50, color='green')
    sns.lineplot(x = df.columns[3:],y= co2_per_capita, label='CO2 per capita', marker='o')
    sns.scatterplot(x = df.columns[3:],y= co2_per_capita, s=50, color='red')
    for x, y in zip(df.columns[3:], foreign):
        ax.text(x, y + y * 0.02, f'{y:,.2f}', ha='center', fontsize=8)
    for x, y in zip(df.columns[3:], co2_per_capita):
        ax.text(x, y + y * 0.02, f'{y:,.2f}', ha='center', fontsize=8)

    ax.set_title(f"{country_name}'s Foreign Direct Investment & CO2 per Capita")
    plt.legend()
    plt.xlabel('Year')
    plt.xticks(rotation=45)

    ax = plt.subplot(2,2,3)
    sns.lineplot(x = df.columns[3:] , y= co2_total, label='Total CO2 emissions', markers='o' )
    sns.scatterplot(x = df.columns[3:] , y= co2_total,  s=50, color='orange')
    sns.lineplot(x = df.columns[3:], y =total_greenhouse_gas, label='Total Green Hourse Gas', marker='o')
    sns.scatterplot(x = df.columns[3:], y =total_greenhouse_gas, color='purple', s=50)
    for x, y in zip(df.columns[3:], co2_total):
        ax.text(x, y + y * 0.02, f'{y:,.2f}', ha='center', fontsize=8)
    for x, y in zip(df.columns[3:], total_greenhouse_gas):
        ax.text(x, y + y * 0.02, f'{y:,.2f}', ha='center', fontsize=8)
    plt.legend()
    plt.xticks(rotation=45)
    plt.xlabel('Year')
    ax.set_title(f"{country_name}'s CO2 Emissions & Greenhouse Gases")
    
    ax=plt.subplot(2,2,4)
    sns.lineplot(x = df.columns[3:], y = trade, label='Trade', marker='o')
    sns.scatterplot(x=df.columns[3:], y = trade, s=50, color='brown')
    for x, y in zip(df.columns[3:], trade):
        ax.text(x, y + y * 0.02, f'{y:,.2f}', ha='center', fontsize=8)
    ax.set_title(f"{country_name}'s Trade over years")
    plt.xlabel('Year')
    plt.ylabel('Trade (% of GDP')
    plt.xticks(rotation=45)
    plt.legend()

    # Then, add spacing between subplots
    plt.tight_layout()
    plt.subplots_adjust(
        hspace=0.4,  # vertical spacing between rows
        wspace=0.3   # horizontal spacing between columns
    )
    plt.savefig('sample.jpg')
    plt.show()

    # print(foreign_direct_invest)
    
```

```python
def plot_for_countries(df, field, countries):
    filt = df['Series Name'] == field
    plt.figure(figsize=(12, 6))

    for country in countries:
        # Extract data
        data = df[filt].loc[country].iloc[3:]
        years = data.index
        values = data.values

        # Plot line
        plt.plot(years, values, label=country)

        # Plot scatter
        sns.scatterplot(x=years, y=values, s=50, color='brown')

        # Add value labels to each scatter point
        for x, y in zip(years, values):
            # v = y / 1e9  # Convert to billions (optional)
            plt.text(x, y, f'{y:.1f}', ha='center', va='bottom', fontsize=7)

    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(True)
    plt.xlabel('Year')
    plt.ylabel(field)
    plt.title('Top CO2 emitted countries')
    plt.savefig('top_co2_countries.jpg', dpi=300)
    plt.show()

plot_for_countries(new_df, 'Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e)', ['United States', 'China', 'Russian Federation', 'Japan', 'India', 'Germany', 'United Kingdom', 'Canada'])
```

```python
def bar_of_field(df, country, field):
    country_code = df.loc[country, 'Country Code'].iloc[0]
    
    filt = (df['Country Code'] == country_code) & (df['Series Name'] == field)
    data = df[filt].iloc[0, 3:]
    
    
    ax = data.plot(kind='bar', figsize=(12, 6), color='skyblue', edgecolor='black')
    plt.title(f"{country}'s CO2 emissions overtime", fontsize=14, weight='bold')
    plt.ylabel('CO2 emissions', fontsize=12)
    plt.xlabel('Year', fontsize=12)
    plt.xticks(rotation=45)
    
    # add value labels on top of each bar
    for i, bar in enumerate(ax.patches):
        value = bar.get_height()
        if field == 'GDP (current US$)':
            value_in_billions = value / 1e9 # convert to billions
            text = f'{value_in_billions:.2f}B'
        else:
            value_in_billions = value
            text = f'{value_in_billions:.2f}'
            
        # print(value)
        ax.text(bar.get_x() + bar.get_width()/ 2, value*0.95,
        text ,
        ha='center', va='top', fontsize=8, rotation=90, color='black')
    
    plt.tight_layout()
    plt.savefig('us_co2.jpg')
    plt.show()
```

```python
def average_change(df, country, field):
    country_code = df.loc[country, 'Country Code'].iloc[0]
    filt = (df['Country Code'] == country_code) & (df['Series Name'] == field)
    return (new_df[filt].iloc[0, 3:]).diff().mean()
```
